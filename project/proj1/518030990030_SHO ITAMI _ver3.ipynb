## 数据预处理

```python
import numpy as np
from sklearn.model_selection import train_test_split

def label(data):
    y_data = data['price_range'].apply(lambda x: 0 if x in [0, 1] else 1)
    X_data = data.drop('price_range', axis=1)
    return X_data, y_data

def split(X, y):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
    X_test, X_val, y_test, y_val = train_test_split(X_test, y_test, test_size=0.5, random_state=1)
    return X_train, X_test, X_val, y_train, y_test, y_val

# 归一函数，离散函数
from sklearn.preprocessing import MinMaxScaler, KBinsDiscretizer
def normalization(data):
    n = MinMaxScaler()
    n.fit(data)
    return pd.DataFrame(n.transform(data))

def discretization(data):
    d = KBinsDiscretizer(n_bins=10, encode='ordinal', strategy='uniform')
    return pd.DataFrame(d.fit_transform(data))
```

## 模型实现

### Naive Bayes

```python
class NavieBayesClassifer():
    def __init__(self):
        super().__init__()
        
    def fit(self, X_train, y_train):
        X = np.array(X_train)
        y = np.array(y_train)
        self.parameters = {}
        self.classes = np.unique(y)
        for cls in self.classes:
            index_list = X[np.where(y == cls)]
            mean = np.mean(index_list, axis=0, keepdims=True)
            var = np.var(index_list, axis=0, keepdims=True)
            prior = index_list.shape[0]/X.shape[0]
            self.parameters[str(cls)] = {"mean": mean, "var": var, "prior": prior}
    
    def pdf(self, X, cls):
        eps = 1e-15
        mean = self.parameters[str(cls)]["mean"]
        var = self.parameters[str(cls)]["var"]

        numerator = np.exp(-(np.array(X) - mean) ** 2 / (2 * var + eps))
        denominator = np.sqrt(2 * np.pi * var + eps)

        result = np.sum(np.log(numerator / denominator), axis=1, keepdims=True)
        return result.T
    
    def predict(self, X_data):
        output = []
        for cls in self.classes:
            prior = np.log(self.parameters[str(cls)]["prior"])
            posterior = self.pdf(X_data, cls)
            prediction = prior + posterior
            output.append(prediction)
            
        output = np.reshape(output, (self.classes.shape[0], X_data.shape[0]))
        prediction = np.argmax(output, axis=0)
        return prediction
```

### Support Vector Machine

```python
from sklearn.svm import SVC

class SVMClassifer():
    def __init__(self):
        super().__init__()
        self.svm = SVC(C=5, kernel='rbf', decision_function_shape='ovr')
              
    def fit(self, X, y):
        self.svm.fit(X, y)
    
    def predict(self, X):
        return self.svm.predict(X)
```

### Logistic Regression

```python
def log_loss(y_true, y_pred, eps=1e-15):
    p = np.clip(y_pred, eps, 1-eps)
    loss = np.sum(- y_true * np.log(p) - (1 - y_true) * np.log(1-p))
    return loss / len(y_true)

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

class LogisticRegressionClassifer():
    def __init__(self, lr=.1, iteration=5000, report_interval=50, threshold =1e-4, report = True):
        super().__init__()
        self.lr = lr
        self.iteration = iteration
        self.interval = report_interval 
        self.threshold = threshold
        self.report = report
        self.iteration_list = []
        self.train_loss_list = []
        self.val_loss_list = []
        self.maintain_time = 10
        
    def initialize_weights(self, features):
        limit = np.sqrt(1 / features)
        w = np.random.uniform(-limit, limit, (features, 1))
        b = 0
        self.w = np.insert(w, 0, b, axis=0)
    
    def add_valid_parameter(self, X_val, y_val):
        self.X_val = X_val
        self.y_val = np.reshape(np.array(y_val), (np.array(y_val).shape[0], 1))
        self.add_val = True

    def fit(self, X, y):
        X = np.array(X)
        y = np.array(y)
        
        samples, features = X.shape
        self.initialize_weights(features)
        
        X = np.insert(X, 0, 1, axis=1)
        y_true = np.reshape(y, (samples, 1))

        # 梯度训练
        for _iter in range(self.iteration):
            y_pred = np.round(sigmoid(np.dot(X, self.w))).astype(int)
            self.w = self.w - self.lr * np.dot(X.T, y_pred - y_true)
            
            if _iter % self.interval == 0:
                train_loss = log_loss(y_true, y_pred)
                val_loss = log_loss(self.y_val, self.predict(self.X_val))
                self.train_loss_list.append(train_loss)
                self.val_loss_list.append(val_loss) 
                
                if self.report:
                    self.iteration_list.append(_iter)
                
                if len(self.val_loss_list)>= self.maintain_time and np.var(self.val_loss_list[-self.maintain_time:]) < self.threshold:
                    break

    def predict(self, X_data):
        X = np.insert(np.array(X_data), 0, 1, axis=1)
        y_pred = np.round(sigmoid(np.dot(X, self.w)))
        return y_pred.astype(int)
```

## 模型测试
在给定的数据集上使用所实现的模型进行训练和预测
数据采用ROC曲线计算面积AUC来衡量模型的性能指标
当AUC越接近1.0，检测方法真实性越高

```python
# 测试函数
import time
from matplotlib import pyplot as plt
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc, roc_auc_score

def train_and_test(features, labels, classfier, add_val = False, predict_train = False):

    X_train, X_test, X_val, y_train, y_test, y_val = split(features, labels)
    
    if add_val == True:
        classfier.add_valid_parameter(X_val, y_val)
    

    start = time.time()
    classfier.fit(X_train, y_train)
    end = time.time()
    train_time = (end - start) * 1000
   
    pred_train = classfier.predict(X_train)
    true_train = y_train
    
    # 预测test值
    start = time.time()
    predict = classfier.predict(X_test)
    end = time.time()
    test_time = (end - start) * 1000
    
    true = y_test
    
    if predict_train == True:
        return train_time, test_time, predict, true, pred_train, true_train
    else:
        return train_time, test_time, predict, true
    
def plotRUC(y_true, y_pred, title=None):
    f_pos, t_pos, thresh = roc_curve(y_true, y_pred)
    auc_area = auc(f_pos, t_pos)

    plt.plot(f_pos, t_pos, 'darkorange', lw=2, label='AUC = %.2f' % auc_area)
    plt.legend(loc='lower right')
    plt.plot([0, 1], [0, 1], color='navy', linestyle='--')
    plt.title('%s' % title)
    plt.ylabel('True Positive Rate')
    plt.xlabel('False Positive Rate')
    plt.show()
    
def report(y_true, y_pred):
    print(classification_report(y_true, y_pred, target_names=['low price', 'high price']))
```

```python
# NaiveBayes Model
features, labels = label(data)
features = discretization(features)
classfier = NavieBayesClassifer()
train_time_nb, test_time_nb, predict_nb, true_nb = train_and_test(features, labels, classfier)
report(true_nb, predict_nb)
plotRUC(true_nb, predict_nb, title='Navie Bayes')
```

```python
# SVM Model
features, labels = label(data)
svm_classfier = SVMClassifer()
train_time_svm, test_time_svm, predict_svm, true_svm = train_and_test(features, labels, svm_classfier)
print(report(true_svm, predict_svm))
plotRUC(true_svm, predict_svm, title='SVM')
```

```python
# Logistic Regression Model
features, labels = label(data)
features = normalization(features)
lr_classfier = LogisticRegressionClassifer(lr = .1, iteration = 2000, report_interval = 50, report=True)
train_time_lr, test_time_lr, predict_lr, true_lr = train_and_test(features, labels, lr_classfier, add_val=True)
print(classification_report(true_lr, predict_lr))
plotRUC(true_lr, predict_lr, title='Logistic Regression')
```

## 结果分析

```python
def deal(classfier, features, labels, add_val = False):
    total_train_time = []
    total_test_time = []
    train_auc = []
    test_auc = []

    # 多次测验取平均
    for _ in range(20):
        train_time, test_time, predict, true, train_pred, train_true = train_and_test(features, labels, classfier, add_val, predict_train = True )
        total_train_time.append(train_time)
        total_test_time.append(test_time)
        train_auc.append(roc_auc_score(train_pred, train_true))
        test_auc.append(roc_auc_score(predict, true))
    
    return np.mean(total_train_time),np.mean(total_test_time), np.mean(test_auc), np.mean(train_auc)

features, labels = label(data)

pro_features = discretization(features)
nb_classfier = NavieBayesClassifer()
nb_train_time, nb_test_time, nb_test_auc, nb_train_auc = deal(nb_classfier, features, labels, add_val = False)

pro_features = normalization(features)
lr_classfier = LogisticRegressionClassifer(iteration = 2000, report_interval = 100, report = False)
lr_train_time, lr_test_time, lr_test_auc, lr_train_auc = deal(lr_classfier, pro_features, labels, add_val = True)

svm_classfier = SVMClassifer()
svm_train_time, svm_test_time, svm_test_auc, svm_train_auc = deal(svm_classfier, features, labels, add_val = False)

index=['training time(ms) ','test time(ms)','training accuracy ','test accuracy ']
res_table=pd.DataFrame({
    'Naive Bayes': pd.Series([nb_train_time, nb_test_time, nb_train_auc, nb_test_auc],index=index),
    'SVM': pd.Series([svm_train_time, svm_test_time, svm_train_auc, svm_test_auc],index=index),
    'Logistic Regression': pd.Series([lr_train_time, lr_test_time, lr_train_auc, lr_test_auc],index=index),
})
print(res_table)

modals = ['NaiveBayes', 'Logistic Regression', 'SVM']
accu = [nb_test_auc, lr_test_auc, svm_test_auc]

plt.bar(modals, accu, color=['lightblue', 'pink', 'lightgrey'])
plt.ylabel("accuracy rate")
plt.title("the accuracy of three models")
plt.show()

timeNeed = [nb_train_time, lr_train_time, svm_train_time]

plt.bar(modals, timeNeed, color=['lightblue', 'pink', 'lightgrey'])
plt.ylabel("time for training")
plt.title("the perfomance of training")
plt.show()
```

## 各模型的优缺点：

1. NaiveBayes

##### 优点：
* 训练时间短，训练过程简单，速度快
##### 缺点：
* 准确率没有其他模型高
* 样本属性若存在关联，其效果会降低


2. SVM
##### 优点：
* 鲁棒性好，增、删非支持向量样本对模型没有影响、支持向量样本集具有一定的鲁棒性、对核的选取不敏感
* 支持非线性问题

##### 缺点：
* 解决多分类问题存在困难
* 对缺失数据敏感
* 对于核函数的高维映射解释力不强，尤其是径向基函数
    
3. Logitstic Regression
##### 优点：
* 模型清晰，背后的概率推导经得住推敲
* 参数代表每个特征对输出的影响，可解释性强

##### 缺点：
* 容易产生欠拟合
* 分类精度不高
* 特征空间很大时，性能不好
